# Machine Learning

## Training and Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
df_train.columns
```

```python
X = df_train.iloc[:,:-1]
y = df_train.iloc[:,-1]
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=101)
```

```python
X_train
```

```python
y_train
```

## Modelling

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
# models
models = {
    "LinearRegression": LinearRegression(),
    "LogisticRegression": LogisticRegression(),
    "RandomForestClassifier": RandomForestClassifier(),
    "DecisionTree": DecisionTreeClassifier(),
    "knn": KNeighborsClassifier()
}

# Function to fit and score models
def fit_score(models, X_train, X_test, y_train, y_test):
    # Set random seed
    np.random.seed(42)
    # Make a dictionary to keep model scores
    model_scores = {}
    # Loop through models
    for name, model in models.items():
        # Fit the model to the data
        model.fit(X_train, y_train)
        # Evaluate the model and append its score to model_scores
        # Here name is the ML model name and "model" is actual ML model
        model_scores[name] = round((model.score(X_test, y_test))*100,3)
    return model_scores
```

```python
model_score = fit_score(models = models,
                           X_train = X_train,
                           X_test = X_test,
                           y_train = y_train,
                           y_test = y_test)

model_score
```

```python
# Comparing the model
model_compare = pd.DataFrame(model_score, index=["accuracy"])
model_compare.T.plot.bar();
```

We can see that KNeighboursClassifier() performed better with accuracy score of 83%

```python
kn = KNeighborsClassifier()
```

```python
# Fitting the score
kn.fit(X_train, y_train)
```

```python
y_pred = kn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
acc = accuracy_score(y_test, y_pred)
acc
```

## Hyperparametertuning using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV 
```

```python
# Different hyperparameters for our Logistic Regression model
log_reg_grid = {"n_neighbors": [5, 20, 30],
               "weights": ["uniform", "distance"],
               "leaf_size": [30, 100, 10]}

# Setup grid hyperparameter search for LogisticRegression
gs_log_reg = GridSearchCV(KNeighborsClassifier(),
                         param_grid=log_reg_grid,
                         cv=5,
                         verbose=True)

# Fit grid hyperparameter search model
gs_log_reg.fit(X_train, y_train)
```

```python
gs_log_reg.best_params_
```

```python
# Evaluate grid search KNeighboursClassifier model
new_acc = gs_log_reg.score(X_test, y_test)
print("The accuracy score after doing hyperaparametertuning is:", round(new_acc*100,3),"%")
```

```python
pred = gs_log_reg.predict(X_test)
```

```python
# Classification Report
print(classification_report(y_test, pred))
```

```python
sns.heatmap(confusion_matrix(y_test, pred), annot=True);
```
